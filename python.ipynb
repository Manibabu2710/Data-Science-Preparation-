```python
import matplotlib.pyplot as plt
```

```python
x=np.array([3,7,4,2,5,1,8])
```

```python
y=x**2+2
```

```python
print(x,y)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(y,x,label="Horse Power")
plt.legend()
plt.title("Usage With AGE")
plt.xlabel("AGE")
plt.ylabel("Power")
plt.show()
```
